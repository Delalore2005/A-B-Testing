A - n пользователей, х успешны
В - m пользователей, y пользователей успешны

p_A = x/n
p_B = y/m

H_0 p_A = p_B = p_0

p_A - p_B ~ N(0, g^2)

D(p_A - p_B) = D(p_a) + D(p_b)

D(p_a) = p_0 * (1 - p_0) / n
D(p_b) = p_0 * (1 - p_0) / m

SE = sqrt(p_0 * 1 - p_0) * (1/n + 1/m)

Z = (p_a - p_b) / SE

```python
alpha = 0.05
```

```python
def find_p(conv_cm, orders_cm):
    p_A = conv_cm['A'][-1]
    p_B = conv_cm['B'][-1]
    n_A = vis_cum['A'][-1]
    n_B = vis_cum['B'][-1]
    p = (orders_cm['A'][-1] + orders_cm['B'][-1]) / (n_A + n_B)
    z = (p_B - p_A) / np.sqrt(p * (1 - p) * (1 / n_A + 1 / n_B))
    dist = stats.norm(0,1)
    p_v = (1- dist.cdf(abs(z)))*2
    return p_v
```

```python
find_p(conv, orders_cum)
```
